# 天猫用户重复购买预测

- 商家通常在节假日进行大规模促销
- 面向特定目标用户投放才能最大化收益
- 对每个店铺，“老顾客”是他们最希望争取的，因此想发优惠券给他们

# 数据描述

# 预测评价指标

$$ 
\text{AUC}=\frac{\sum_{i \in \text{positive Class}} \text{rank}_i - \frac{M(1+M)}{2}}{M*N}
$$
